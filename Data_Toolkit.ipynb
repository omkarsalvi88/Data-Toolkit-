1.  What is NumPy, and why is it widely used in Python?

2.  How does broadcasting work in NumPy?
    - NumPy is a Python library used for fast numerical computations. It provides efficient support for multi-dimensional arrays and mathematical functions, making it essential for data science, machine learning, and scientific computing.Broadcasting in NumPy allows operations on arrays of different shapes by automatically expanding the smaller array to match the shape of the larger one. This works if their dimensions are compatible, meaning one of them is 1 or both are equal. If not, NumPy raises an error. Broadcasting avoids the need for manual reshaping and makes array operations more concise and efficient.


3. What is a Pandas DataFrame?

It consists of rows and columns, where each column can have a different data type (e.g., integers, strings, floats). DataFrames make it easy to load, view, filter, and analyze structured data, making them a core tool in data analysis and manipulation with Python.


4.  Explain the use of the groupby() method in Pandas
    - The groupby() method in Pandas is used to split a DataFrame into groups based on one or more columns, allowing you to perform operations like aggregation, filtering, or transformation on each group separately. It is commonly used for tasks like calculating sums, means, or counts for each group. For example, df.groupby('Category').sum() will group the data by the 'Category' column and then sum the values within each group.

6. What are the differences between NumPy arrays and Python lists?
   -  NumPy arrays and Python lists both store collections of data, but they differ in several key ways. NumPy arrays are more memory-efficient and allow fast mathematical operations on entire arrays, while Python lists are more flexible but slower for numerical tasks. NumPy arrays support only one data type per array and are stored in contiguous memory, making them suitable for numerical computing. In contrast, Python lists can store mixed data types and are more general-purpose but less efficient for large-scale data processing.

Heatmaps are commonly used in correlation analysis, confusion matrices in machine learning, and to visualize large datasets with relationships between variables.


8. What does the term “vectorized operation” mean in NumPy?
   - In NumPy, a vectorized operation refers to performing operations on entire arrays without using explicit loops. These operations are applied element-wise and are optimized internally using fast, low-level implementations, making them much faster and more efficient than looping through elements in Python. For example, adding two arrays with a + b is a vectorized operation that adds each corresponding element directly.

17. Explain the difference between apply() and map() in Pandas?
    - In Pandas, apply() is used to apply a function along an axis of a DataFrame (rows or columns) or on a Series, allowing more complex operations on entire rows, columns, or Series. map() is mainly used with a Series to map values from a dictionary, Series, or function element-wise, typically for simpler value-to-value transformations. In short, apply() works on whole rows/columns or Series and can handle more complex functions, while map() is for element-wise mapping on Series only.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
    - NumPy’s array slicing is faster than Python’s list slicing because NumPy arrays store data in contiguous blocks of memory with a fixed data type, allowing efficient, low-level access and operations using optimized C code. When you slice a NumPy array, it creates a view of the original data without copying it, which saves time and memory. In contrast, Python lists store references to objects scattered in memory, so slicing involves creating new lists with copied references, making it slower and less memory-efficient.

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#13. Create a 3D scatter plot using Plotly.


import plotly.graph_objects as go
import numpy as np

np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=6,
        color=z,
        colorscale='Viridis',
        opacity=0.8
    )
)])


fig.update_layout(
    title='3D Scatter Plot Example',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Show the plot
fig.show()
```
